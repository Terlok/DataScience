```python
import pandas as pd
import urllib.request
from datetime import datetime
import numpy as np
import requests
from datetime import date
year1 = 1990
year2 = 2022
```

```python
NewIndex = {1:24, 2:25, 3:5, 4:6, 5:27, 6:23, 7:26, 8:7, 9:11, 10:13, 11:14, 12:15, 13:16, 14:17, 15:18, 
            16:19, 17:21, 18:22, 19:8, 20:9, 21:10, 22:1, 23:3, 24:2, 25:4,  26:12, 27:20 }
names = ['Вінницька_обл', 'Волинська_обл', 'Дніпропетровська_обл', 'Донецька_обл', 'Житомирська_обл', 'Закарпатська_обл', 'Запорізька_обл', 
         'Івано-Франківська_обл', 'Київська_обл', 'Кіровоградська_обл', 'Луганська_обл', 'Львівська_обл', 'Миколаївська_обл', 'Одеська_обл', 
         'Полтавська_обл', 'Рівенська_обл', 'Сумська_обл', 'Тернопільська_обл', 'Харківська_обл', 'Херсонська_обл', 'Хмельницька_обл', 
         'Черкаська_обл', 'Чернівецька_обл', 'Чернігівська_обл', 'Республіка Крим', 'Київ', 'Севастополь']
province = {i+1: '' for i in range(1, 28)}
```

```python
def save_files(id):
  link = f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={NewIndex[id]}&year1={year1}&year2={year2}&type=Mean"
  time = datetime.now().strftime("%d.%m.%Y %H:%M:%S")
  filename = f"№{id}: {names[id-1]} ({time}).csv"
  province.update({id: filename})
  open(filename, 'wb').write(requests.get(link).content)
  print(f"File: '{filename}' uploaded successfully!")

for i in range(1, 28):
  save_files(i)
```

```python
def get_data():
  with open('ResultTable.csv', 'w', encoding='utf-8') as Resultfile:
    for i in range(27):
      with open(f'{province[i+1]}', 'r', encoding='utf-8') as Somefile:
        buffer = Somefile.readlines()
        for j in range(len(buffer)):
          buffer[j] = ''.join(buffer[j][:-2].split())+f',{i+1}'
        buffer[1] = 'Year,Week,SMN,SMT,VCI,TCI,VHI,Region'
        buffer[2] = buffer[2].split('>')[2]
      print(*buffer[2:-1], sep='\n', file=Resultfile)
      with open(f'{province[i+1]}', 'w', encoding='utf-8') as Somefile:
        print(*buffer[1:-1], sep='\n', file=Somefile)
  print('Parsing was completed!')

get_data()

headers = ['Year', 'Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI', 'Region']
DataSet = pd.read_csv('ResultTable.csv', names = headers)
DataSet = DataSet.drop(DataSet.loc[DataSet['VHI'] == -1].index)
```

```python
def Extreme_points_min(region, year):
  Values = min(DataSet[(DataSet.Year == year) & (DataSet.Region == region)]['VHI'])
  return DataSet[(DataSet.VHI == Values) & (DataSet.Year == year) & (DataSet.Region == region)]
```

```python
def Extreme_points_max(region, year):
  Values = max(DataSet[(DataSet.Year == year) & (DataSet.Region == region)]['VHI'])
  return DataSet[(DataSet.VHI == Values) & (DataSet.Year == year) & (DataSet.Region == region)]
```

```python
def Extreme_droughts(region):
  return DataSet[(DataSet.VHI <= 15) & (DataSet.Region == region)]
```

```python
def Moderate_drought(region):
  return DataSet[(DataSet.VHI <= 35) & (DataSet.VHI > 15) & (DataSet.Region == region)]
```
